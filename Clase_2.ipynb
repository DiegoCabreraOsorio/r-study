```r
# izquierda
ggplot(data = millas) +
  geom_point(mapping = aes(x = cilindrada, y = autopista))

# derecha
ggplot(data = millas) +
  geom_smooth(mapping = aes(x = cilindrada, y = autopista))
```

Puedes establecer la forma para un punto, pero no puedes establecer la “forma” de una línea. Por otro lado, para una línea podrías elegir el tipo de línea (linetype). geom_smooth() dibujará una línea diferente, con un tipo de línea distinto (linetype), para cada valor único de la variable que asignes al tipo de línea (linetype)

```r
summary(datos::millas)
```

```r
ggplot(data = millas) +
  geom_smooth(mapping = aes(x = cilindrada, y = autopista, linetype = traccion))

ggplot(data = millas) +
  geom_smooth(mapping = aes( x= cilindrada, y=autopista, linetype = clase))
```

```r
??mapping
```

También podemos hacer múltiples "ploteos" dentro de un mismo gráfico, para ello se procede de la siguiente manera:

```r
ggplot(data = millas, mapping = aes(x = cilindrada, y = autopista)) +
  geom_point() +
  geom_smooth()
```

```r
ggplot(data = millas, mapping = aes(x = cilindrada, y = autopista)) +
  geom_point(mapping = aes(color = clase)) +
  geom_smooth()
```

El siguiente gráfico, es lo que en Python conocemos como "boxplot":

###Ejercicio
*coord()* usa coordenadas polares. Convierte un gráfico de barras apiladas en un gráfico circular usando *coord_polar()*.

```r
bar <- ggplot(data = diamantes) +
  geom_bar(
    mapping = aes(x = corte, fill = corte),
    show.legend = FALSE,
    width = 1
  ) +
  theme(aspect.ratio = 1) +
  labs(x = NULL, y = NULL)

bar + coord_flip()
bar + coord_polar()
```

Con "<-" asignamos un único nombre a nuestra "función", es como que llamamos *bar* al gráfico que hicimos, luego le podemos aplicar el cambio a coordenadas, flipeo, polares, etc.

```r
bar <- ggplot(data = diamantes) +
  geom_bar(mapping = aes(x=corte))
```
