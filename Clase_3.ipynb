###Añadir nuevas variables con mutate()


mutate() siempre agrega nuevas columnas al final de un conjunto de datos, así que comenzaremos creando un conjunto de datos más pequeño para que podamos ver las nuevas variables. Recuerda que cuando usas RStudio, la manera más fácil de ver todas las columnas es View().

````r
vuelos_sml <- select(vuelos,
  anio:dia, #selecciono desde la columna anio hasta la columna dia
  starts_with("atraso"),  #luego, las que comienzan con la palabra "atraso"
  distancia,  #agrego la columna distancia y la de tiempo de vuelo
  tiempo_vuelo
)
mutate(vuelos_sml,  #filamnente muto, es decir, agrego nuevas columnas!
  ganancia = atraso_salida - atraso_llegada,
  velocidad = distancia / tiempo_vuelo * 60
)
````

````r
por_destino <- group_by(vuelos, destino) #agrupo mi dataframe por destinos
atraso <- summarise(por_destino,  #y ahora, resumo, de acuerdo a cada destino:
  conteo = n(),   #contaré los vuelos que hay
  distancia = mean(distancia, na.rm = TRUE),  #la distancia promedio de cada vuelo a dicho destino
  atraso = mean(atraso_llegada, na.rm = TRUE)   #atraso promedio a cada destino
)
atraso <- filter(atraso, conteo > 20, destino != "HNL")   #filtro por atrasos, donde haya más de 20 y que no vayan a honululu

ggplot(data = atraso, mapping = aes(x = distancia, y = atraso)) +
  geom_point(aes(size = conteo), alpha = 1/3) +
  geom_smooth(se = FALSE)
````
